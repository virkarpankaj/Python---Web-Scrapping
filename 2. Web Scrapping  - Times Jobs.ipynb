```python
import requests

from bs4 import BeautifulSoup

import pandas as pd

```

```python
# step1 : Get the HTML
# step2 : Parse the HTML

url  = ' https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation='

response = requests.get(url)

htmlContent = response.content  # response.text

soup = BeautifulSoup(htmlContent, 'lxml' )  # optional -'html.parser' -- dont write 'lxml.parser'

print(soup.prettify)
```

```python
# accessing single job on that page 

job = soup.find('li',class_ = 'clearfix job-bx wht-shd-bx')
print(job)

# the below info belongs to only 1 job
```

```python
# finding inside the job Properties

company_name = job.find('h3',class_ = 'joblist-comp-name').text.strip()

skill = job.find('span', class_ = 'srp-skills').text.strip().replace(' ','')

posted_date = job.find('span',class_ = 'sim-posted').text.strip()

print(f'Company Name is \n{company_name} \n' )
print(f'Skills Required are \n({skill}) \n')
print(f'Job Posted - {posted_date}\n')
```

```python
# accessing all the jobs on that page - By iterating

jobs = soup.find_all('li',class_ = 'clearfix job-bx wht-shd-bx')

for job in jobs:
    
    company_name = job.find('h3',class_ = 'joblist-comp-name').text.strip()

    skill = job.find('span', class_ = 'srp-skills').text.strip().replace(' ','')

    posted_date = job.find('span',class_ = 'sim-posted').text.strip()
    
    print(f'Company Name is === {company_name} \n' )
    print(f'Skills Required are === ({skill}) \n')
    print(f'Job Posted === {posted_date}\n')
    
    print('')
```

```python
# more info added 


jobs = soup.find_all('li',class_ = 'clearfix job-bx wht-shd-bx')

for job in jobs:
     
    more_info = job.header.h2.a['href']
    
    print(more_info)
    print('')
```

```python
# filtering out the old jobs wrt Date of posting

jobs = soup.find_all('li',class_ = 'clearfix job-bx wht-shd-bx')

for job in jobs:
    
    posted_date = job.find('span',class_ = 'sim-posted').text.strip()
    
    if 'few' in posted_date:  
        company_name = job.find('h3',class_ = 'joblist-comp-name').text.strip()

        skill = job.find('span', class_ = 'srp-skills').text.strip().replace(' ','')

        print(f'Company Name is === {company_name} \n' )
        print(f'Skills Required are === ({skill}) \n')
        print(f'Job Posted === {posted_date}\n')

        print('')
```

```python
# filtering based on skills

print(' Mention the Unfamiliar Skill ')

unfamiliar_skill = input('>')

print(f'Filtering out {unfamiliar_skill}')

jobs = soup.find_all('li',class_ = 'clearfix job-bx wht-shd-bx')

for job in jobs:
    
    posted_date = job.find('span',class_ = 'sim-posted').text.strip()
    if 'few' in posted_date:
        
        skill = job.find('span', class_ = 'srp-skills').text.strip().replace(' ','')
        company_name = job.find('h3',class_ = 'joblist-comp-name').text.strip()
        more_info = job.header.h2.a['href']
    
        if unfamiliar_skill not in skill:
            print(f'Company Name is === {company_name} \n' )
            print(more_info)
            print(f'Skills Required are === ({skill}) \n')
            print(f'Job Posted === {posted_date}\n')

            print('')
```

```python
# def a function

# Function 1:

def get_soup(url):
    
    response = requests.get(url)

    htmlContent = response.content  # response.text

    soup = BeautifulSoup(htmlContent, 'lxml' )  # optional -'html.parser' -- dont write 'lxml.parser'

    print(soup.prettify)
```
